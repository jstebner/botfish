```python
from chessboard2fen.boardDetection import ChessboardDetector
from chessboard2fen.chessboard import Chessboard
from glob import glob
import cv2
```

```python
detect = ChessboardDetector(
    "chessboard2fen/models/detection", "chessboard2fen/models/classification.h5"
)
# board = Chessboard()
filenames = sorted(glob("chessboard2fen/input_imgs/*"))
```

```python
game = list()
for file in filenames:
    img = cv2.imread(file)
    corners = detect.predict_board_corners(img)
    if len(corners) == 4:
        predictions = detect.predictBoard(img, corners)
        
    game.append(predictions)
        
```

```python
fenChar = ["p", "P", "q", "Q", "k", "K", "e", "b", "B", "n", "N", "r", "R"]
for idx, pos in enumerate(game):
    print(idx+1)
    translate = [fenChar[cell] for cell in pos]
    for i in range(8):
        for j in range(8):
            print(translate[i*8 + j], end=' ')
        print()
    print('\n')
```
